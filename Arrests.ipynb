1. Load the dataset
First, we open the dataset and import our packages. 

```python
import pandas as pd
import matplotlib.pyplot as plt

arrests_data = pd.read_csv('Arrests_file.csv')



arrests_data.head()
```

2. Group the data by neighborhood and count arrests
Next, we group the data by neighborhood and count the number of arrests.



```python
neighborhood_arrest_counts = arrests_data['INCIDENTNEIGHBORHOOD'].value_counts()
neighborhood_arrest_counts.head()
```

3. Sort neighborhoods by arrest counts
Now, we sort the neighborhoods by the number of arrests in ascending order to find the neighborhoods with the least crime.

```python
neighborhood_arrest_counts = neighborhood_arrest_counts.sort_values()
neighborhood_arrest_counts.head
```

4. Display all neighborhoods sorted by arrest counts
Here, we display all neighborhoods in ascending order of arrest counts.

```python
print("Neighborhoods and Arrest Counts (Sorted):")
print(neighborhood_arrest_counts)
```

5. Create a bar chart to visualize the data
Finally, we create a bar chart to show the number of arrests per neighborhood.

```python
plt.figure(figsize=(20, 10))
plt.bar(neighborhood_arrest_counts.index, neighborhood_arrest_counts.values, color='skyblue')


plt.xlabel('Neighborhoods', fontsize=14)
plt.ylabel('Number of Arrests', fontsize=14)
plt.title('Arrests by Neighborhood', fontsize=13)


plt.xticks(
    ticks=range(len(neighborhood_arrest_counts)),
    labels=neighborhood_arrest_counts.index,
    rotation=90,
    ha='right',
    fontsize=10
)


plt.tight_layout()


plt.show()
```
