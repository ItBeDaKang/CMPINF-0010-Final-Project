1. Load the dataset
First, we open the dataset and import our packages. 

```python
import pandas as pd
import matplotlib.pyplot as plt

arrests_data = pd.read_csv('Arrests_file.csv')



arrests_data.head()
```

2. Group the data by neighborhood and count arrests
Next, we group the data by neighborhoods and count the number of arrests in each.



```python
#arrests_per_neighborhood = arrests_data.groupby('INCIDENTNEIGHBORHOOD').size().reset_index(name='Arrest Count')


neighborhood_arrest_counts = arrests_data['INCIDENTNEIGHBORHOOD'].value_counts()
neighborhood_arrest_counts.head()
```
